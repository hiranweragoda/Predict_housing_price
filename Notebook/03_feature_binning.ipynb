```python
import numpy as np
import pandas as pd
```

### 02. Import Dataset

```python
df = pd.read_csv("data/processed/estate_missing_values_handled.csv")
df.head()
```

### 03. Feature Binning

```python
def property_age_binning(age):
    if age <= 15:
        return 'New'
    elif age <= 35:
        return 'Moderate'
    else:
        return 'Old'
```

```python
df['PropertyAge_bins'] = df['PropertyAge'].apply(property_age_binning)
del df['PropertyAge']
```
